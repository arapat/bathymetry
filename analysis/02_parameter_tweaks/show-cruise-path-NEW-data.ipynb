# Show cruise path
The idea here is to plot cruise paths with scores and visualize the scoring in a rough sense. Obviously, it's better to have the scores overlain on bathymetry, but that's what we do with pycmeditor..

```python
import numpy as np
import io
%matplotlib inline
import matplotlib.pyplot as plt
import pygmt
```

```python
filename = "/swot2/features/cm_data/public/US_multi2/NBP1210.cm.edit.all.baseparam"
data = np.loadtxt(filename, usecols=(1,2,-1))
```

```python
region = [
    data[:,0].min() - 1,
    data[:,0].max() + 1,
    data[:,1].min() - 1,
    data[:,1].max() + 1,
]

region[0] = -115
region[1] = -80
region[2] = -74
print(region)
```

```python
fig = pygmt.Figure()
fig.basemap(region=region, projection="M15c",frame=True)
fig.coast(land="black", water="skyblue")
pygmt.makecpt(cmap="polar", series=[0.0, 1.0])
fig.plot(
    x=data[:,0],
    y=data[:,1],
    color=data[:,2],
    cmap=True,
    style='p'
)
fig.colorbar(frame='af+l"Score"')
fig.show()
```
