## Show cruise path
The idea here is to plot cruise paths with scores and visualize the scoring in a rough sense. Obviously, it's better to have the scores overlain on bathymetry, but that's what we do with pycmeditor..

```python
import numpy as np
import io
```

```python
filename = "/swot2/features/cm_data/public/US_multi/NBP0002_ed.cm.edit.US_multi"
```

```python
# load the lon, lat, score
data = np.loadtxt(filename, usecols=(1,2,5,8))
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import cm
```

```python
polar = cm.get_cmap('bwr')
colors=polar(data[:,3])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.scatter(data[:,0],data[:,1],c=colors)
```

```python
colors_flag = []
for j,flag in enumerate(data[:,2]):
    if flag == 9999: #blue= flag
        colors_flag.append('#1f77b4')
    elif flag == 9998: # green = repeat
        colors_flag.append('#33ff36')
    else: # red = good
        colors_flag.append('#d62728')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.scatter(data[:,0],data[:,1],c=colors_flag)
ax.legend()
```
