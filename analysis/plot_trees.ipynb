## Train-test split model
Model trees for the US_multi train-test split. This model includes all features.

```python
filename = "/swot2/features/MODEL_RUNS/00_ALL_FEATURES/train-test-split/runtime_models/US_multi_model.pkl"
with open(filename,"rb") as f:
    gbm = pickle.load(f)
gbm.save_model('model.temp') # I put this step in there to manually edit the feature labels, couldn't figure out how to change them
```

```python
mymodel = lightgbm.Booster(model_file='model.temp')
lightgbm.create_tree_digraph(mymodel,tree_index=0)
```
