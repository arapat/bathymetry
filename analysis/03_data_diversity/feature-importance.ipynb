# Feature Importance -- training data diversity
The models trained so far have been from a single source or a combination of only two sources. The results have not been great. By training a model from all data sources, we increase the diversity of inputs. This might sacrifice performance on a per-source basis, but the model is generalized, which is what we need in practice.

The following models have 10 features removed (up through NDP30m). Boosting parameters are set to address overfitting:

    1. extra_trees = true
    2. feature_fraction = 0.5
    3. bagging_fraction = 0.5
    4. bagging_freq = 10
    5. lambda_l2 = 1.0
    6. max_bin = 127
    7. min_data_in_leaf = 5000
    8. learning_rate = 0.05
    9. rounds = 500
