# Score distribution -- Data diversity
What follows are the cumulative distributions of the scores for each run. For "bad" samples, the score is plotted in blue from 1 to 0. For "good" samples, the score is plotted in red from 0 to 1. So, for a perfect classifier, you would basically have a blue peak at 0, a red peak at 1.

## All sources Model, tested on each data source


```python
sources = ['AGSO','JAMSTEC','JAMSTEC2','NGA','NGA2','NGDC','NOAA_geodas','SIO','US_multi']
base_dir = "/swot2/features/MODEL_RUNS/14_ALL_PARAMS/all_sources_model/runtime_scores"
```

```python
for source in sources:
    path = os.path.join(base_dir, "model_all_test_{}_scores.pkl".format(source,source))
    with open(path, 'rb') as f:
        data = pickle.load(f)
    plot_scores(data, source)
```

## Individual models
Here is the same train/test split but with a model trained on single data sources.

```python
base_dir = "/swot2/features/MODEL_RUNS/14_ALL_PARAMS/all_sources_model/runtime_scores"
for source in sources:
    path = os.path.join(base_dir, "model_{}_test_{}_scores.pkl".format(source,source))
    with open(path, 'rb') as f:
        data = pickle.load(f)
    plot_scores(data, source)
```
